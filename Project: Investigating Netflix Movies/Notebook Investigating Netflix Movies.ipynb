```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like

listd = []

# Corrected the filtering of the DataFrame
netflix_df_90s_movies = netflix_df[(netflix_df["release_year"] >= 1990) & 
                                   (netflix_df["release_year"] <= 1999) & 
                                   (netflix_df["type"] == "Movie")]

# Corrected the conversion to list
ld = netflix_df_90s_movies["duration"].tolist()

# Iterate over the list
for du in ld:
    listd.append(ld.count(du))

m = max(listd)

pro = listd.index(m)

duration = ld[pro]

print("the most frequent movie duration in the 90's is " + str(duration))
```

```python
# Creating a sub-dataframe with entries from the 90s and action 
netflix_90s_action_df = netflix_df[(netflix_df["release_year"] >= 1990) & (netflix_df["release_year"] <= 1999) & (netflix_df["genre"] == "Action") &( netflix_df["type"]=="Movie"  )  ]

lact = netflix_90s_action_df["duration"].tolist()

short_movie_count = 0
for i in lact:
    if i < 90:
        short_movie_count += 1  # Fixed the increment operator
    else:
        continue  # Added a continue statement to handle the else case

print(short_movie_count)
```

```python
# Printing the result of the project
result_text = f"""
The most frequent movie duration in the 90's is {duration} minutes.
There are {short_movie_count} action movies from the 90s with a duration of less than 90 minutes.
"""

result_text
```
